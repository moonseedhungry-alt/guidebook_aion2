```python
import time
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager

def collect_nc_guide_urls(start_id, end_id):
    print(f"üöÄ ÏàòÏßë ÏãúÏûë: CategoryId {start_id} ~ {end_id}")

    options = webdriver.ChromeOptions()
    # options.add_argument('--headless') # Ïûò ÎêòÎäîÍ±∞ ÌôïÏù∏ÌïòÏãúÎ©¥ Ï£ºÏÑù Ìï¥Ï†úÌïòÏÑ∏Ïöî (ÏÜçÎèÑ Ìñ•ÏÉÅ)
    options.add_argument("--window-size=1920,1080")
    options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36")
    
    service = Service(ChromeDriverManager().install())
    driver = webdriver.Chrome(service=service, options=options)
    
    all_urls = set()

    try:
        for cat_id in range(start_id, end_id + 1):
            url = f"https://aion2.plaync.com/ko-kr/guidebook/list#categoryId={cat_id}"
            print(f"\nüîÑ [Category {cat_id}] Ïù¥Îèô Ï§ë...")
            
            driver.get(url)

            try:
                # 1. Î™ÖÌôïÌïú ÌÅ¥ÎûòÏä§Î™ÖÏù¥ Î°úÎî©Îê† ÎïåÍπåÏßÄ ÎåÄÍ∏∞
                wait = WebDriverWait(driver, 10)
                target_class = "ncgbg-guide-depth-2-guide-item-link"
                
                wait.until(EC.presence_of_element_located((By.CLASS_NAME, target_class)))
                time.sleep(1) # Î†åÎçîÎßÅ ÏïàÏ†ïÌôî

                # 2. Ìï¥Îãπ ÌÅ¥ÎûòÏä§Î•º Í∞ÄÏßÑ Î™®Îì† ÏöîÏÜåÎ•º Ï∞æÏùå
                elements = driver.find_elements(By.CLASS_NAME, target_class)
                
                count = 0
                for el in elements:
                    # SeleniumÏù¥ hrefÎ•º Í∞ÄÏ†∏Ïò¨ Îïå ÏûêÎèôÏúºÎ°ú http://... Ï†ÑÏ≤¥ Í≤ΩÎ°úÎ°ú Î≥ÄÌôòÌï¥Ï§çÎãàÎã§.
                    full_url = el.get_attribute("href")
                    
                    # Í∞ÄÎÅî Îπà ÎßÅÌÅ¨ÎÇò Ï§ëÎ≥µÏù¥ ÏûàÏùÑ Ïàò ÏûàÏúºÎãà ÌïÑÌÑ∞ÎßÅ
                    if full_url and "view?title=" in full_url:
                        all_urls.add(full_url)
                        count += 1
                
                print(f"   ‚úÖ {count}Í∞ú ÏàòÏßë ÏÑ±Í≥µ")

            except Exception:
                print(f"   ‚ö†Ô∏è [Category {cat_id}] Ìï¥Îãπ Ïπ¥ÌÖåÍ≥†Î¶¨Ïóê Í∏ÄÏù¥ ÏóÜÍ±∞ÎÇò Î°úÎî© Ïã§Ìå®")
                continue

    except Exception as e:
        print(f"‚ùå ÏóêÎü¨ Î∞úÏÉù: {e}")
    
    finally:
        driver.quit()

    return list(all_urls)

# --- Ïã§Ìñâ ---

    # 4234 ~ 4244 Ï†ÑÏ≤¥ ÏàòÏßë
final_urls = collect_nc_guide_urls(4234, 4244)

print(f"\nüéâ ÏµúÏ¢Ö ÏàòÏßë Í≤∞Í≥º: Ï¥ù {len(final_urls)}Í∞ú")
for url in final_urls:
    print(url)
```

```python
import os
import asyncio
from dotenv import load_dotenv
from langchain_community.document_loaders import PlaywrightURLLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_openai import OpenAIEmbeddings
from langchain_pinecone import PineconeVectorStore

# ÌôòÍ≤Ω Î≥ÄÏàò Î°úÎìú
load_dotenv()

INDEX_NAME = "aino2-guide-rag"

final_urls = [
    "https://aion2.plaync.com/ko-kr/guidebook/view?title=%EA%B2%80%EC%84%B1%20%EC%8A%A4%ED%82%AC",
    # ... ÎÇòÎ®∏ÏßÄ URL ...
]

# [ÌïµÏã¨] ÌäπÏ†ï ÏòÅÏó≠(.ncgbt-article)Ïùò ÌÖçÏä§Ìä∏Îßå Ï∂îÏ∂úÌïòÎäî Ïª§Ïä§ÌÖÄ Ìï®Ïàò
def get_article_text(page):
    # .ncgbt-article ÌÅ¥ÎûòÏä§Î•º Í∞ÄÏßÑ ÏöîÏÜåÎ•º Ï∞æÏùå
    locator = page.locator(".ncgbt-article")
    
    # Ìï¥Îãπ ÏöîÏÜåÍ∞Ä ÏûàÏúºÎ©¥ ÌÖçÏä§Ìä∏ Î∞òÌôò, ÏóÜÏúºÎ©¥ Îπà Î¨∏ÏûêÏó¥ Î∞òÌôò (ÏóêÎü¨ Î∞©ÏßÄ)
    if locator.count() > 0:
        return locator.first.inner_text()
    else:
        return ""

async def main():
    print(f"1. [{len(final_urls)}]Í∞úÏùò URLÏóêÏÑú Îç∞Ïù¥ÌÑ∞ Î°úÎî© Ï§ë... (Playwright)")
    
    loader = PlaywrightURLLoader(
        urls=final_urls,
        continue_on_failure=True,
        # remove_selectors ÎåÄÏã† evaluatorÎ•º ÏÇ¨Ïö©ÌïòÏó¨ ÏõêÌïòÎäî Î∂ÄÎ∂ÑÎßå Í∞ÄÏ†∏Ïò¥
        evaluator=get_article_text
    )
    
    docs = await loader.aload()
    
    # ÎÇ¥Ïö©Ïù¥ ÏóÜÎäî Î¨∏ÏÑú(Îπà Î¨∏ÏûêÏó¥) ÌïÑÌÑ∞ÎßÅ
    docs = [doc for doc in docs if doc.page_content.strip()]
    
    print(f"-> Î°úÎìú ÏôÑÎ£å! Ïú†Ìö® Î¨∏ÏÑú Í∞úÏàò: {len(docs)}")

    if not docs:
        print("Î¨∏ÏÑúÍ∞Ä Î°úÎìúÎêòÏßÄ ÏïäÏïòÏäµÎãàÎã§. Ï¢ÖÎ£åÌï©ÎãàÎã§.")
        return

    # ÌôïÏù∏Ïö©: Ï≤´ Î≤àÏß∏ Î¨∏ÏÑú ÏïûÎ∂ÄÎ∂Ñ Ï∂úÎ†•
    print(f"\n[ÏàòÏßë Í≤∞Í≥º ÏÉòÌîå]\n{docs[0].page_content[:300]}...\n")

    print("2. ÌÖçÏä§Ìä∏ Î∂ÑÌï† Ï§ë...")
    text_splitter = RecursiveCharacterTextSplitter(
        chunk_size=1000,
        chunk_overlap=200
    )
    
    splits = text_splitter.split_documents(docs)
    print(f"-> Î∂ÑÌï† ÏôÑÎ£å! Ï¥ù Ï≤≠ÌÅ¨ Ïàò: {len(splits)}")

    print(f"3. Pinecone('{INDEX_NAME}')Ïóê Îç∞Ïù¥ÌÑ∞ Ï†ÄÏû• ÏãúÏûë...")
    
    embeddings = OpenAIEmbeddings(model="text-embedding-3-small")

    PineconeVectorStore.from_documents(
        documents=splits,
        embedding=embeddings,
        index_name=INDEX_NAME
    )
    
    print("üéâ Î™®Îì† ÏûëÏóÖÏù¥ ÏÑ±Í≥µÏ†ÅÏúºÎ°ú ÏôÑÎ£åÎêòÏóàÏäµÎãàÎã§!")

if __name__ == "__main__":
    asyncio.run(main())
```
